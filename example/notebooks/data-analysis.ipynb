# Example notebook for data analysis

This is an example notebook, showing how to add a figure to the article using a label

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#| label: fig-histogram

# Generate 100 random data points, normally distributed around 10 with a standard deviation of 2
data = np.random.normal(10, 2, 100)

# Create a histogram of the data
plt.hist(data, bins=15, color='blue', alpha=0.7)

# place curvefit over histogram
mu, sigma = 10, 2
x = np.linspace(0, 20, 100)
plt.plot(x, 100 * (1/(sigma * np.sqrt(2 * np.pi))) * np.exp( - (x - mu)**2 / (2 * sigma**2) ), color='red')

plt.title('Histogram of Random Data with Curve Fit')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
